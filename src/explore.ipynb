```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import tree
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report

import pickle
from pickle import dump
```

```python
with open('/workspaces/kiliangil-random-forest-prediciendo-diabetes/data/processed/dataset.pkl', 'rb') as f:
    X_train, X_test, y_train, y_test = pickle.load(f)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

#### Prueba 1. Modelado Random Forest con seleción de hiperparámetros manual
Se han ido probando de manera manual distintos parámetros hasta encontrar las mejores métricas.

```python
model = RandomForestClassifier(random_state = 42, n_estimators = 300, criterion = 'entropy', max_depth= 7)

# entrenamiento
model.fit(X_train, y_train)
```

```python
y_pred_test = model.predict(X_test)
y_pred_test
```

```python
y_pred_train = model.predict(X_train)
y_pred_train
```

```python
accuracy_test = accuracy_score(y_test, y_pred_test)
accuracy_train = accuracy_score(y_train, y_pred_train)

f1_score_test = f1_score(y_test, y_pred_test, average='micro')
f1_score_train = f1_score(y_train, y_pred_train, average='micro')

precision_test = precision_score(y_test, y_pred_test, average='micro')
precision_train = precision_score(y_train, y_pred_train, average='micro')

recall_test = recall_score(y_test, y_pred_test, average='micro')
recall_train = recall_score(y_train, y_pred_train, average='micro')

print("Accuracy Test: ", accuracy_test)
print("F1 score Test: ", f1_score_test)
print("Precision Test: ", precision_test)
print("Recall Test: ", recall_test)

print("Accuracy Train: ", accuracy_train)
print("F1 score Train: ", f1_score_train)
print("Precision Train: ", precision_train)
print("Recall Train: ", recall_train)
```

#### Pruebas con distintos hiperparámetros

```python
# Lista para guardar resultados
results = []

# Valores que voy a probar
n_estimators_list = [50, 100, 200, 300, 500]
max_depth_list = [3, 5, 7, 9, None]  # None = sense límit

# Bucle para probar distintas combinaciones de hiperparámetros
for n in n_estimators_list:
    for d in max_depth_list:
        model = RandomForestClassifier(
            random_state=42,
            n_estimators=n,
            criterion='entropy',
            max_depth=d
        )
        model.fit(X_train, y_train)
        y_pred_test = model.predict(X_test)
        y_pred_train = model.predict(X_train)
        acc_test = accuracy_score(y_test, y_pred_test)
        acc_train = accuracy_score(y_train, y_pred_train)
        results.append({'n_estimators': n, 'max_depth': d, 'accuracy_test': acc_test, 'accuracy_train': acc_train})

# Convertir a DataFrame para visualización
results_df = pd.DataFrame(results)
print(results_df)
```

#### Gráficación de resultados

```python
plt.figure(figsize=(10, 6))

for d in max_depth_list:
    subset = results_df[results_df['max_depth'] == d]
    plt.plot(subset['n_estimators'], subset['accuracy_test'], marker='o', linestyle='-', label=f'Test (max_depth={d})')
    plt.plot(subset['n_estimators'], subset['accuracy_train'], marker='x', linestyle='--', alpha=0.6, label=f'Train (max_depth={d})')

plt.title('Impacto de n_estimators y max_depth en la precisión (Train vs Test)')
plt.xlabel('n_estimators')
plt.ylabel('Accuracy')
plt.legend(title='Tipo de dato y max_depth', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18,6), sharey=True)

#Accuracy TEST
sns.barplot(
    data=results_df,
    x='max_depth',
    y='accuracy_test',
    hue='n_estimators',
    palette='Greens',
    errorbar=None,
    ax=axes[0]
)
axes[0].set_title('Precisión en TEST según max_depth y n_estimators')
axes[0].set_xlabel('max_depth')
axes[0].set_ylabel('Accuracy')
axes[0].grid(axis='y')
axes[0].legend(title='n_estimators')

#Accuracy TRAIN
sns.barplot(
    data=results_df,
    x='max_depth',
    y='accuracy_train',
    hue='n_estimators',
    palette='Blues',
    errorbar=None,
    ax=axes[1]
)
axes[1].set_title('Precisión en TRAIN según max_depth y n_estimators')
axes[1].set_xlabel('max_depth')
axes[1].set_ylabel('Accuracy')
axes[1].grid(axis='y')
axes[1].legend(title='n_estimators')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

pivot_train = results_df.pivot(index='max_depth', columns='n_estimators', values='accuracy_train')
pivot_test = results_df.pivot(index='max_depth', columns='n_estimators', values='accuracy_test')

sns.heatmap(pivot_test, annot=True, fmt=".3f", cmap='Greens', ax=axes[0])
axes[0].set_title('Accuracy TEST según n_estimators y max_depth')
axes[0].set_xlabel('n_estimators')
axes[0].set_ylabel('max_depth')

sns.heatmap(pivot_train, annot=True, fmt=".3f", cmap='Blues', ax=axes[1])
axes[1].set_title('Accuracy TRAIN según n_estimators y max_depth')
axes[1].set_xlabel('n_estimators')
axes[1].set_ylabel('max_depth')

plt.tight_layout()
plt.show()
```

#### Conclusiones
El Random Forest mejora claramente las métricas respecto al Decision Tree, gracias a la combinación de múltiples árboles y ajuste de hiperparámetros.
Por un lado, se observa que un número suficiente de árboles (n_estimators ≈ 300) estabiliza la precisión, y una profundidad intermedia (max_depth ≈ 7) permite un buen equilibrio entre train y test. Por otro lado, combinaciones extremas producen overfitting o underfitting.

Por lo tanto, tras las pruebas, el mejor modelo final (n_estimators=300, max_depth=7, criterion='entropy') logra las mejores métricas de test y mantiene la robustez frente al sobreajuste, esto permite comprobar que en la Prueba 1 del modelo con ajuste manual de hiperparámetros, también se haía conseguido llegar a las mejores métricas posibles.

Dicho esto, Random Forest ofrece un modelo más preciso y generalizable que un árbol individual, capturando correctamente la mayoría de casos positivos sin comprometer la estabilidad.
